```python
#loan is approved or not
import numpy as np

# Initializing parameters
w1 = np.random.randn()
w2 = np.random.randn()
w3 = np.random.randn()
b = np.random.randn()
lr = 0.01
epochs = 1400

# Input/output
x1 = np.array([100,150,300,350])  # credit score
x2 = np.array([50000, 40000, 60000, 30000])  # income
x3 = np.array([0, 1, 0, 1])
y  = np.array([0, 1, 0, 1])

#normalization
x1=(x1-np.mean(x1))/np.std(x1)
x2=(x2-np.mean(x2))/np.std(x2)
x3=(x3-np.mean(x3))/np.std(x3)

# Step activation function
def step_activation(y):
    return 1 if y > 0 else 0

# Gradient Descent
def gradient_descent(w1, w2, w3, b, lr, epochs, x1, x2, x3, y):
    n = len(x1)
    for i in range(epochs):
        y_pred = (w1 * x1 + w2 * x2 + w3 * x3) + b
        cost = (1/n) * sum((y - y_pred) ** 2)
        dw1 = (2/n) * sum(x1 * (y - y_pred))
        dw2 = (2/n) * sum(x2 * (y - y_pred))
        dw3 = (2/n) * sum(x3 * (y - y_pred))
        db  = (2/n) * sum(y - y_pred)

        w1 = w1 + lr * dw1
        w2 = w2 + lr * dw2
        w3 = w3 + lr * dw3
        b  = b +lr * db

        if i % 100 == 0:
            print(f"\nEpoch {i}")
            print("w1 =", w1)
            print("w2 =", w2)
            print("w3 =", w3)
            print("b  =", b)
            print("cost =", cost)

    return w1, w2, w3, b

# Train the model
weight_1, weight_2, weight_3, bias = gradient_descent(w1, w2, w3, b, lr, epochs, x1, x2, x3, y)

# Predict for new input: (780, 60000, 1)
input1=(350-np.mean(x1))/np.std(x1)
input2=(60000-np.mean(x2))/np.std(x2)
input3=(1-np.mean(x3))/np.std(x3)
output=(input1*weight_1+input2*weight_2+input3*weight_3)+bias
output1=step_activation(output)
print(output1)
```
