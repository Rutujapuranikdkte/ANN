```python
import numpy as np
import matplotlib.pyplot as plt

# Inputs
x = np.array([1, 2])
y = np.array([5])
weight = np.array([0.5, 0.5])
rate = 0.01
iterations = 300

# To store cost over iterations
errors = []

for i in range(iterations):
    y_pred = np.dot(weight, x)
    error = y - y_pred
    weight = weight + rate * error * x
    errors.append(float(error**2))  # Store squared error for graph

    if i % 100 == 0:
        print(f"Iteration {i} \tError: {error}")

print("Final weights:", weight)

# Plotting error vs. iterations
plt.plot(range(iterations), errors)
plt.title("LMS Error Over Iterations")
plt.xlabel("Iterations")
plt.ylabel("Squared Error")
plt.grid(True)
plt.show()
```
