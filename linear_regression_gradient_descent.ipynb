```python
#linear regression using gradient descent
import numpy as np
import matplotlib.pyplot as plt

x=np.array([1,2,3,4,5])#study hours
y=np.array([50,60,70,80,90])#exam score
lr=0.01
m_curr,b_curr=0,0
n=len(x)
iterations=1200

#gradient descent
for i in range(iterations):
  print("\n")
  print(i)
  y_pred=m_curr*x+b_curr
  cost=(1/n)*sum((y-y_pred)**2)
  dm=(-2/n)*sum(x*(y-y_pred))
  db=(-2/n)*sum(y-y_pred)
  m_curr=m_curr-lr*dm
  b_curr=b_curr-lr*db
  if iterations%100==0:
    print("iteration:",i)
    print("\ncost:",cost)
    print("\tweight:",m_curr)
    print("\tbias:",b_curr)

plt.scatter(x,y,color="blue",label="True Label")
plt.plot(x,m_curr*x+b_curr,color="red",label="fitted line")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression')
plt.legend()
plt.show()



```
