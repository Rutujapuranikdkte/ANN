```python
#normalization
import numpy as np

w1=np.random.rand()
w2=np.random.rand()
b=np.random.rand()
lr=0.01
epoch=450

x1=[600,780]
x2=[30000,60000]
y=[1,0]

x1=(x1-np.mean(x1))/np.std(x1)
x2=(x2-np.mean(x2))/np.std(x2)

def gradient_descent(w1,w2,b,lr,epoch,x1,x2):
  n=len(x1)
  n1=len(x2)

for iterations in range(epoch):
  y_pred=(w1*x1+w2*x2)+b
  cost=(1/n)*sum((y-y_pred)**2)
  dw1=(2/n)*sum(x1*(y-y_pred))
  dw2=(2/n1)*sum(x2*(y-y_pred))
  dw2=(2/n2)*sum(x2*(y-y_pred))
  db=(2/n)*sum(y-y_pred)

  w1-=lr*dw1
  w2=lr*dw2
  b-=lr*db
  print("\nm=",w1,"\nn=",w2,"\nb=","\tcost=",cost)


gradient_descent(w1,w2,b,lr,epoch,x1,x2)
```
