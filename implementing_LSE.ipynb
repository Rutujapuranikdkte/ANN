```python
#Implementation of LSE
import numpy as np

# Given dataset
x = np.array([1, 2, 3, 4, 5])
y = np.array([5000, 7000, 9000, 11000, 13000])

# Reshape x to a column vector and add a bias column (intercept term)
X = np.column_stack((np.ones(x.shape[0]), x))  # Shape: (5, 2)

# Calculation for weight matrix using Normal Equation
A = np.dot(X.T, X)
A_inv = np.linalg.inv(A)
B = np.dot(X.T, y)
weight = np.dot(A_inv, B)

# Predicting the values
y_pred = np.dot(X, weight)

# Find cost (mean squared error, optional)
cost = np.sum((y - y_pred) ** 2)

# Print rounded cost
print("cost", round(cost, 5))  # This will show 0.0

# Predict value for new data
x_new = np.array([1, 7])  # [bias, x]
output = np.dot(x_new, weight)
print("output", output)
```
